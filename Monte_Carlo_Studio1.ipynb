```python
myu1=[]

for i in range(1,200):
    x=np.linspace(0,np.pi,i)
    y=np.sin(x)
    myu1.append(y.sum()/len(y))
    
```

```python
plt.plot(range(1,200),myu1,'r-')
plt.grid()
plt.ylim([-0.1,1.2])
```

```python
import scipy as sp
```

```python
x=np.pi*sp.stats.uniform.rvs(size=100)
```

```python
y=np.sin(x)
y.sum()/len(y)
```

```python
myu=[]

for i in range(1,4000):
    x=np.pi*sp.stats.uniform.rvs(size=i)
    y=np.sin(x)
    myu.append(y.sum()/len(y))
```

```python
plt.plot(range(1,4000),myu,'r--')
plt.plot([0,4000],[2/np.pi,2/np.pi],'b-',lw=3)
plt.grid()
plt.ylim([-0.1,1.2])
```
