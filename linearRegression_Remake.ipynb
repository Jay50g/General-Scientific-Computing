```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn import *
```

```python
x=np.linspace(-10,10,30)

y=30*np.sin(2*x)+np.random.normal(0,2,len(x))
plt.plot(x,y,'ro')
```

```python
f= lambda x,a,b,c : a*np.cos(b*x+c)
```

```python
ans,covar=sp.optimize.curve_fit(f,x,y,[50,1.9,5])
```

```python
ans
```

```python
t=np.linspace(-10,10,400)
yp=f(t,ans[0],ans[1],ans[2])
plt.plot(t,yp,'g-',lw=3)
plt.plot(x,y,'ro')
```

```python
sm=sp.interpolate.make_smoothing_spline(x,y,lam=0.0005)
smD=sm.derivative()
smD2=smD.derivative()
```

```python
plt.plot(x,y,'ro')
plt.plot(t,sm(t),'g-',lw=3)
plt.plot(t,smD(t),'r-',lw=3)
plt.plot(t,smD2(t),'k-',lw=3)
```

```python
inter=sp.interpolate.make_interp_spline(x,y,k=5)
interD=inter.derivative()
```

```python
plt.plot(x,y,'ro')
plt.plot(t,inter(t),'g-',lw=3)
plt.plot(t,interD(t),'r-',lw=3)
```

```python
cubic=sp.interpolate.CubicSpline(x,y)
cubicD=cubic.derivative()
```

```python
plt.plot(x,y,'ro')
plt.plot(t,cubic(t),'g-',lw=3)
plt.plot(t,cubicD(t),'r-',lw=3)
```

```python
x=np.linspace(0.1,10,25)
t=np.linspace(0,10,500)
y=10-5/(0.5+x)+np.random.standard_normal(len(x))
plt.plot(x,y,'ro')
```

```python
z=np.polyfit(x,y,2)
p=np.poly1d(z)
yp=p(t)
p
```

```python
plt.plot(x,y,'ro')
plt.plot(t,yp,'g-',lw=3)
```

```python
ans,covar=sp.optimize.curve_fit(f2,x,y)
```
