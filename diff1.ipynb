```python
y=sym.symbols('y',cls=sym.Function)
x=sym.symbols('x')
```

```python
eq2=sym.Eq(y(x).diff(x),(6*x**5-18)/x**3)
```

```python
ans=sym.dsolve(eq2,y(x),ics={y(1):6})
```

```python
ans
```

```python
eq3=sym.Eq(y(x).diff(),15*x**4-6*x)
```

```python
ans2=sym.dsolve(eq3,y(x),ics={y(-1):1})
ans2
```

```python
sym.plot(ans2.rhs,eq3.rhs,(x,-1,2))
```
