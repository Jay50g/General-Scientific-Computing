```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,5000)
```

```python
px=np.exp(-1/2*x**2)/np.sqrt(2*np.pi)
```

```python
plt.plot(x,px,'r-')
```

```python
y=np.cos(x)
```

```python
plt.plot(x,y,'g-')
```

```python
f=y*px
```

```python
plt.plot(x,f,'b-')
```

```python
digInt=sp.integrate.trapezoid(f,x)
```

```python
X=np.random.standard_normal(20000)
```

```python
fR=np.cos(X)
```

```python
fMC=[]
for i in range(len(X)):
    fMC.append(fR[:i+1].mean())
```

```python
plt.plot(range(20000),digInt*np.ones(20000),'r-',lw=3)
plt.plot(range(20000),fMC,'b-',lw=3)
```
