```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
a=np.array([1,0,0]*4)
```

```python
a
```

```python
A1=np.vstack((a,np.roll(a,1),np.roll(a,2)))
```

```python
A1
```

```python
s=np.array([480,650,580,390])
```

```python
A2=A1*np.repeat(s,3)
```

```python
A2
```

```python
a1=np.zeros(12)
a1[:3]=1
a1
```

```python
A3=np.vstack((a1,np.roll(a1,3),np.roll(a1,6),np.roll(a1,9)))
A3
```

```python
A=np.vstack((A1,A2,A3))
A
```

```python
masses=np.array([10,16,8])
spaces=np.array([6800,8700,5300])
goods=np.array([18,15,23,12])
b=np.hstack((masses,spaces,goods))
```

```python
b
```

```python
money=np.array([3100,3800,3500,2800])
c=np.repeat(money,3)
c
```

```python
ans=sp.optimize.linprog(-c,A,b)
```
