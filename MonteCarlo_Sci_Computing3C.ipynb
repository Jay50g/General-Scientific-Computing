```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#x=sp.stats.uniform.rvs(size=200)*np.pi
```

```python
sam=[]
for i in range(1,200):
    x=sp.stats.uniform.rvs(size=i)*np.pi
    y=np.sin(x)
    sam.append(y.sum()/i)  
```

```python
plt.plot(sam,'r.')
plt.plot([0,200],[2/np.pi]*2,'b-',lw=3)
plt.ylim((-0.2,1.2))
```

```python
x=sp.stats.uniform.rvs(size=200)*np.pi
y=np.sin(x)
```

```python
timeSeries=np.zeros(200)
for i in range(200):
    timeSeries[i]=y[:i+1].sum()/(i+1)
```

```python
plt.plot(timeSeries,'r-')
plt.ylim((-0.2,1.2))
plt.plot([0,200],[2/np.pi]*2,'b-',lw=3)
```
