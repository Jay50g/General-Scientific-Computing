```python
import  numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
## sample the X space in a determininstic way
```

```python
x=np.linspace(0,np.pi,1000)
y=np.sin(x)
```

```python
y.sum()/1000
```

```python
2/np.pi
```

```python
average=[]

for i in range(1,2000):
    x=np.linspace(0,np.pi,i)
    y=np.sin(x)
    average.append(y.sum()/len(y))
     
```

```python
plt.plot(range(1,2000),average,'ro-')
```

```python
### Select values of random variable X with a uniform probability
x1=sp.stats.uniform.rvs(size=100)*np.pi
y1=np.sin(x1)
y1.sum()/100
```

```python
ave1=[]

for i in range(1,2000):
    x1=sp.stats.uniform.rvs(size=i)*np.pi
    y1=np.sin(x1)
    ave1.append(y1.sum()/i)
    
    
```

```python
plt.plot(range(1,2000),ave1,'r.')
plt.plot(range(1,2000),average,'b-',lw=3)
plt.ylim([-0.1,1.2])
```
