```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name
        self.age=age
        self.__gender=gender


    def change_name(self,x):   # setter
        self.__name=x

    def show_name(self):       # reader 
        return self.__name

    def spend_a_year(self):
        self.age+=1


    def eat(self):
        if self.__gender=='Male':
            x='Rice'
            y='He'
        else: 
            x='Noodles'
            y='She'
        print(y, ' eats ',x)


    def spend_a_day(self):
        print('Plays video games')
```

```python
class Student(Person):
    def __init__(self,name,age,gender,school):
        super().__init__(name,age,gender)
        self.school=school


    def spend_a_day(self):
        print('Studies Math')

    def eat(self):
        super().eat()
        print('Will go to study after the meal')


    
```
