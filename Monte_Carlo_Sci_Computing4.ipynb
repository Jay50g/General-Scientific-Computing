```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,np.pi,20)
y=np.sin(x)
t=np.linspace(0,np.pi,1000)
plt.plot(x,y,'ro')
plt.plot(t,np.sin(t),'b-')
```

```python
y.sum()/len(y)
```

```python
samp1=[]
for i in range(1,200):
    x=np.linspace(0,np.pi,i)
    y=np.sin(x)
    samp1.append(y.sum()/len(y))
```

```python
plt.plot(samp1,'m-',lw=5)
plt.ylim([-0.2,1.2])
plt.plot([0,200],[2/np.pi,2/np.pi],'g-',lw=3)
```

```python
x2=sp.stats.uniform.rvs(size=100)*np.pi
y2=np.sin(x2)
y2.sum()/len(y2)
```

```python
samp2=[]
for i in range(1,200):
    x=sp.stats.uniform.rvs(size=i)*np.pi
    y=np.sin(x)
    samp2.append(y.sum()/i)
```

```python
plt.plot(samp2,'r.')
plt.ylim([-0.2,1.2])
plt.plot(samp1,'m-',lw=5)
plt.plot([0,200],[2/np.pi,2/np.pi],'g-',lw=3)
```

```python
x=sp.stats.uniform.rvs(size=200)*np.pi
y=np.sin(x)
```

```python
timeSeries=np.zeros(200)
```

```python
for i in range(200):
    timeSeries[i]=y[:i+1].sum()/(i+1)
```

```python
plt.plot(timeSeries,'r-')
plt.plot([0,200],[2/np.pi,2/np.pi])
plt.ylim([-0.2,1.2])
```
