```python
samples=[]
for i in range(1,2000):
    x=np.linspace(0,np.pi,i)
    y=np.sin(x)
    samples.append(y.sum()/i)
    
```

```python
plt.plot(range(1,2000),np.array(samples),'ro')
plt.plot([0,1999],[2/np.pi,2/np.pi],'b-',lw=3)
```

```python
x=sp.stats.uniform.rvs(size=100)*np.pi
```

```python
y=np.sin(x)
y.sum()/len(x)
```

```python
samp=np.zeros(599)
for i in range(1,600):
    x=sp.stats.uniform.rvs(size=i)*np.pi
    samp[i-1]=np.sin(x).sum()/i
```

```python
plt.plot(samp,'r-')
plt.plot(samples[:200],'mo')
plt.plot([0,600],[2/np.pi,2/np.pi],'b-',lw=3)
plt.ylim([-0.2,1.2])
```

```python
x=sp.stats.uniform.rvs(size=1000)*np.pi
```

```python
y=np.sin(x)
```

```python
y1=np.zeros(len(y))
for i in range(1,len(y)+1):
    y1[i-1]=(y[:i]).sum()/i
```

```python
plt.plot(y1,'r-')
plt.ylim([-0.2,1.2])
plt.plot([0,1000],[2/np.pi,2/np.pi])
```
